# Mapping the Magnetic Field of a Coil
## Purpose
We will measure the magnetic field produced by a current carrying coil of wire and compare it to the theoretical result obtained from the Biot-Savart Law.
## Introduction
In 1820 the theoretical model for current carrying wires was elucidated by Jean-Baptiste Biot and Félix Savart. The law follows from the idea that moving charges in a conductor create a magnetic field. By adding up infinitesimal lengths of current along the conductor, the contributions to the total magnetic field can be integrated for any particular geometry. The increment of magnetic field from a small length $dL$ of conductor is

$$d\vec{B}=\frac{\mu_o I d\vec{L}\times\hat{r}}{4\pi r^2}$$

where $\mu_o$ is the vacuum permeability, $I$ is the current and $r$ describes the distance away from the conductor where we are interested in knowing the magnetic field. The hatted version of $r$ describes the unit vector pointing toward our position of interest.

The cross product can be simplified if we only are concerned with magnitude of the magnetic field. We could in principle keep track of the direction from the cross-product.

$$d\left|\vec{L}\times\hat{r}\right| = dL \sin\theta$$

Our integrand becomes

$$dB=\frac{\mu_o I dL \sin\theta}{4\pi r^2}$$

If we are interested in the field along the axis going through our loop of wire, we should simplify using the geometry involved. Looking at Fig. 1 below, we can simplify our integrand.
<center>
Fig. 1. The geometry of a current carrying loop of wire when considering the magnetic field along the central axis.</center>

We can redefine the sin term.

$$\sin\theta = \frac{R}{r}$$

We can also redefine the distance $r$

$$r = \left(R^2 + z^2\right)^{1/2}$$

Plugging these substitutions into $dB$ we get

$$dB=\frac{\mu_o I R dL}{4\pi r^3}$$

$$dB=\frac{\mu_o I R dL}{4\pi \left(R^2+z^2\right)^{3/2}}$$

Now, we only need to integrate both sides to get the total magnetic field, $B$. Since the only integrand that matters on the right hand side is $dL$, this integral is very easy. We integrate $dL$ around the loop to get the circumference of the loop.

$$\int dB=\int \frac{\mu_o I R}{4\pi \left(R^2+z^2\right)^{3/2}} dL$$

$$\int dB=\frac{\mu_o I R}{4\pi \left(R^2+z^2\right)^{3/2}} \int dL$$

$$B =\frac{\mu_o I R}{4\pi \left(R^2+z^2\right)^{3/2}} 2\pi R$$

$$B = \frac{\mu_o I R^2}{2 \left(R^2+z^2\right)^{3/2}}$$

With this theoretical result we can measure the magnetic field of a current carrying loop and compare. In our experiment we will use coils of wire with 200 and 500 loops. Each loop adds equally to the magnetic field to make it 200 or 500 times stronger than the equation above. For $N$ turns of wire in the coil, we have

$$B = N \frac{\mu_o I R^2}{2 \left(R^2+z^2\right)^{3/2}}$$

That uncertainty does a lot for us at the small distances. This is mainly because the gain setting is really for larger values. We are close to the upper end of the next higher gain setting, and so our data is just a little too large for higher gain setting, putting us at the lower end of the proper gain setting and close to the accuracy of it. For the larger distances, the error bars do not help us much, i.e., they do not overlap with the theoretical curve.

Let's try propagating the error based upon our ability to measure the position. We will do this using the calculus method. If we assume that our measurements have uncertainty, we can propagate that uncertainty as long as we know the theoretical formula we are expecting our data to follow. In this case, our formula is

$$B = N \frac{\mu_o I R^2}{2 \left(R^2+z^2\right)^{3/2}}$$

Uncertainty (or error) comes from the slope of this function relative to any of the measured variables. For example, if the only measured variable were $z$, the slope of $B$ with respect to $z$ defines the change of the function if we were to mis-measure $z$. This idea is related to the Taylor Series expansion of a function. The function $B$ is defined at a value $z$, and we can predict nearby values of the function using the first derivative of the function times the distance away from $z$ that we want to know.

$$B\left(z\right) = B\left(z_o\right) + \frac{dB}{dz}\left(z-z_o\right)\cdots$$

In our experiment, we think we are measuring $z_o$ and $z-z_o$ is the uncertainty of our measurement $z_o$. We often call this $\Delta z$ when calculating uncertainty. In this experiment we actually measure $I$, $R$, and $z$. We could include $N$ too, but we will skip that. So, we need to calculate the uncertainty contributed by our uncertainty in all three of these variables. This turns the Taylor Series into a series of three partial derivatives because we care about how the function behaves in all measured dimensions $I$, $R$, and $z$. The first term in the above Taylor Series is the measured value. The uncertainty is the second term. Our uncertainty in three measurement variable dimensions becomes

$$\Delta B = \frac{\partial B}{\partial I}\Delta I + \frac{\partial B}{\partial R}\Delta R + \frac{\partial B}{\partial z}\Delta z$$

Taking the derivatives, we get

$$\Delta B = \frac{N \mu_o R^2}{2 \left(R^2+z^2\right)^{3/2}}\Delta I + \left|\frac{\left(2N \mu_o I R\right)\left(2 \left(R^2+z^2\right)^{3/2}\right)}{2 \left(R^2+z^2\right)^{3}} - \frac{\left(N \mu_o I R^2\right)\left(6R \left(R^2+z^2\right)^{1/2}\right)}{2 \left(R^2+z^2\right)^{3}}\right|\Delta R + \frac{6Nz\mu_o I R^2}{4 \left(R^2+z^2\right)^{5/2}}\Delta z$$

These things get complicated pretty fast. It is a good idea to do a calculation on paper first. Then, we will implement it in Jupyter and have the results added as a column to our dataframe.

```python
#Define our variables
N200 = 200 #200 turns
N500 = 500 #500 turns
muo = 4*np.pi*1e-7 #physical constant
R = 0.1 #radius of coil
DR = 0.01 #experimental uncertainty in coil radius
I = 1 # current in coil
DI = 0.1 #experimental uncertainty in current
Dz = 0.01 #uncertainty in measurement of distance along the axis

#Pre-caclulate some repeated math
distance_sq = N200Data['distance'] * N200Data['distance']
pythag = distance_sq + R*R

#Calculate each term in the DB equation for easier troubleshooting
term1_N200 = N200*muo*R*R / (2*np.power(pythag, 3/2))*DI
term2_N200 = np.absolute((2*N200*muo*I*R)/(np.power(pythag, 3/2))-(N200*muo*I*R*R)*(6*R)/(2*np.power(pythag, 2.5)))*DR
term3_N200 = 6*N200*N200Data['distance']*muo*I*R*R/(4*(np.power(pythag, 5/2)))*Dz
                       
N200Data['error_prop'] = term1_N200 + term2_N200 + term3_N200
print(N200Data)
```

```python
#Repeat for 500 turn data
#radius is the same, current is the same
#Pre-caclulate some repeated math
distance_sq500 = N500Data['distance'] * N500Data['distance']
pythag500 = distance_sq500 + R*R

#Calculate each term in the DB equation for easier troubleshooting
term1_N500 = N500*muo*R*R / (2*np.power(pythag500, 3/2))*DI
term2_N500 = np.absolute((2*N500*muo*I*R)/(np.power(pythag500, 3/2))-(N500*muo*I*R*R)*(6*R)/(2*np.power(pythag500, 2.5)))*DR
term3_N500 = 6*N500*N500Data['distance']*muo*I*R*R/(4*(np.power(pythag500, 5/2)))*Dz
                       
N500Data['error_prop'] = term1_N500 + term2_N500 + term3_N500
print(N500Data)
```

```python
#And we can adjust spacing
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=0.5)

#Let's plot both sets of data using subplots
plt.subplot(2, 1, 1)
ax1 = plt.gca()
N200Data.plot(kind='scatter',x='distance',y='Measured B',color='red',yerr='error_prop', ax=ax1, label='Measured')
N200Data.plot(kind='line',x='distance',y='Theory B',color='blue', ax=ax1, label='Theory')
plt.title('N = 200 Turns')
plt.xlabel('Distance (m)')
plt.ylabel('Magnetic Field (T)')
plt.axis([-0.01, 0.35, 0, 0.0015])

plt.subplot(2, 1, 2)
ax2 = plt.gca()
N500Data.plot(kind='scatter',x='distance',y='Measured B',color='red',yerr='error_prop', ax=ax2, label='Measured')
N500Data.plot(kind='line',x='distance',y='Theory B',color='blue', ax=ax2, label='Theory')
plt.title('N = 500 Turns')
plt.xlabel('Distance (m)')
plt.ylabel('Magnetic Field (T)')
plt.axis([-.010, 0.35, 0, 0.0035])
plt.show()
```

Those error bars look a bit more reasonable. They don't go off-scale at the small $z$. They seem to make the $N$=200 turn coil data agree with the theory. The $N$=500 turns data still doesn't agree with the theoretical model within the uncertainty.

## Conclusions
The Biot-Savart Law appears to predict the magnetic field along the axis of a circular coil of wire. Our $N$=200 turn data agrees within the experimental uncertainty when considering the uncertainty due to the the measurements of length and current. The $N$=500 turns data is not in agreement within the experimental uncertainty. All of the data is lower than predicted by the theoretical model. This indicates some systematic uncertainty that may be due to the taring of the sensor or the sensor's measurement, itself. The curvature of the Biot-Savart equation follows the data well. Further experiments should identify the source of systematic error and eliminated it to give better agreement.
